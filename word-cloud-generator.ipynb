```python
def load_stopwords_from_file(file_path):
    """
    Loads stopwords from a text file.
    Each word should be on a new line.
    """
    with open(file_path, 'r', encoding='utf-8') as file:
        stopwords = set(word.strip() for word in file)
    return stopwords

# Example usage
persian_stop_words_file = os.path.join(cwd,'STOP WORDS ๐ซ','stop-words.txt')
persian_stop_words = load_stopwords_from_file(persian_stop_words_file)

print(persian_stop_words)
```

```python
def get_text(file_name):
    file_path = os.path.join(cwd,'ghazal-ha-text',file_name)
    with open(file_path, 'r', encoding='utf-8') as file:
        texts = file.read()
        print(f'Text retrieved for {file_name}')
        return texts
```

```python
import os

def save_wordcloud(wordcloud, file_name):
    """Saves a word cloud image to a specified file.

    Args:
        wordcloud: The word cloud object.
        file_name: The desired filename for the output image.
        output_dir: The directory where the image should be saved.
    """

    try:
        output_dir = os.path.join(cwd,'word_clouds')
        # Create the output directory if it doesn't exist
        os.makedirs(output_dir, exist_ok=True)

        # Construct the output path
        save_path = os.path.join(output_dir, file_name)

        # Save the word cloud image
        wordcloud.to_file(save_path)
        print(f"Word cloud saved to {save_path}")
    except Exception as e:
        print(f"Error saving word cloud: {e}")
```

```python
font_path = os.path.join(cwd,'fonts','Vazirmatn-Bold.ttf')
```

```python
def write_top_words(reshaped_freq, num_words=10):
    # Sort the dictionary by values in descending order
    sorted_freq = sorted(reshaped_freq.items(), key=lambda x: x[1], reverse=True)

    # Reverse the order of the words
    for i, (word, freq) in enumerate(sorted_freq[:num_words], start=1):
        reversed_word = word[::-1]
        print(f"{i}. {reversed_word}: {freq}")

```

```python

def generate_persian_wordcloud(text,file_name):
    # Preprocess the text
    words = re.findall(r'\w+', text)
    word_counts = Counter(words)
    
    # Remove stop words
    for word in persian_stop_words:
        if word in word_counts:
            del word_counts[word]
    
    # Create a dictionary of word frequencies
    word_freq = {word: count for word, count in word_counts.items()}
    
    # Reshape the text to handle right-to-left display
    reshaped_words = [arabic_reshaper.reshape(word) for word in word_freq.keys()]
    display_words = [get_display(word) for word in reshaped_words]
    
    # Create a dictionary with reshaped words
    reshaped_freq = dict(zip(display_words, word_freq.values()))
    write_top_words(reshaped_freq, num_words=10)    # Generate word cloud
    wordcloud = WordCloud(
        font_path=font_path,
        background_color='white',
        width=1000,
        height=1000,
        include_numbers=False
    ).generate_from_frequencies(reshaped_freq)
    save_name = file_name.replace('.txt', '') + '.png' 
    # # plt.figure(figsize=(10, 8))
    # # plt.imshow(wordcloud, interpolation='bilinear')
    # # plt.axis('off')
    # # plt.tight_layout(pad=0)
    
    # plt.show()
    save_wordcloud(wordcloud, save_name)
    # Display the word cloud
```

```python
for file in list_of_texts_files:
    texts = get_text(file)
    generate_persian_wordcloud(texts,file)
```
